```python
nominal_cols = ['Loan_ID','Education','Gender','Married','Self_Employed','Property_Area','Loan_Status']

#load_id is unique for each record so it is nominal
# Gender is nominal because there is no order between male and female
# Married is nominal because there is no order between married and unmarried
# Self_Employed is nominal because there is no order between yes and no
# Property_Area is nominal because there is no order between urban, semiurban and rural
# Loan_Status is nominal because there is no order between yes and no

ordinal_cols = ['Dependents','Credit_History']

# Education is ordinal because there is an order between graduate and not graduate graduate > not graduate
# Dependents is ordinal because there is an order between 0,1,2,3+ , 0 < 1 < 2 < 3+
# Credit_History is ordinal because there is an order between 0 and 1 , 0 < 1 0 means no credit history and 1 means has credit history

continue_cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount']

# ApplicantIncome is continuous because it can take any value within a range
# CoapplicantIncome is continuous because it can take any value within a range

Discrete_cols = ['Loan_Amount_Term']

# LoanAmount is discrete because it can take only specific values
# Loan_Amount_Term is discrete because it can take only specific values


print("Nominal columns:", nominal_cols)
print("Ordinal columns:", ordinal_cols)
print("Continuous columns:", continue_cols)
print("Discrete columns:", Discrete_cols)

```

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])

data['Married'] = data['Married'].fillna(data['Married'].mode()[0])

data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])

data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])

data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mode()[0])

data['ApplicantIncome'] = data['ApplicantIncome'].fillna(data['ApplicantIncome'].mean())

data['CoapplicantIncome'] = data['CoapplicantIncome'].fillna(data['CoapplicantIncome'].mean())

data['LoanAmount'] = data['LoanAmount'].fillna(round(data['LoanAmount'].mean(), 1))

data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0])
```
